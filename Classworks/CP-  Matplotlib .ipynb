```python
# continoues errors

from sklearn.gaussian_process import GaussianProcessRegressor as GP

#define the model and draw some data
f= lambda x:x *np.sin(x)
x = np.array([1,3,5,6,8]) #training data
# print(x.shape)
y = f(x) # training target data

#sklearn likes number of samples x number of features
#(5, ) to (5,1)
#compute the gaussian process fit

#so to put it in the right format either use newaxis or reshape
gp = GP().fit(x[:,np.newaxis],y) #gp knows the distribution

#input
xfit=np.linspace(0,10,1000) #test data

yfit,sigma =gp.predict(xfit[:,np.newaxis], return_std = True)
dyfit = 1.96* sigma

plt.plot(x,y,'or')
plt.plot(xfit,yfit,c='grey')

plt.fill_between(xfit,yfit-dyfit,yfit+dyfit,color='grey',alpha= 0.2)
```

### Multiple subplots

```python
# matlab way

# plt.subplot(rows,columns, panels) #set the current plot to this

for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5,str((2,3,i)),fontsize = 18, horizontalalignment = 'center', bbox=dict(facecolor = 'red',alpha = 0.5))
    
```

```python
# Second way - oo stle
fig = plt.figure()
fig.subplots_adjust(hspace=0.4,wspace = 0.4)
for i in range(1,7):
    ax = fig.add_subplot(2,3,i) #return ax and now ax controls that subplot, ax is now the instance of that subplot
    ax.text(0.5,0.5,str((2,3,i)),fontsize = 18, horizontalalignment = 'center', bbox=dict(facecolor = 'red',alpha = 0.5))
```

```python
# 3rd way
x = np.linspace(0,2*np.pi,400)
y = np.sin(x**2)

fig,ax=plt.subplots(2,3,sharex = 'col',sharey='row')
ax[1,0].plot(x,y)
```

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y =np.random.multivariate_normal(mean,cov,3000).T #.T to adj the shape too insert to matplot lib
fig = plt.figure(figsize = (6,6))
grid = plt.GridSpec(4,4,hspace = 0.2,wspace= 0.2)
main_ax = fig.add_subplot(grid[:-1,1:])
y_hist =fig.add_subplot(grid[:-1,0],xticklabels=[],sharey =main_ax)
x_hist =fig.add_subplot(grid[:-1,1:],yticklabels=[],sharex =main_ax)

main_ax.plot(x,y,'ok',markersize=3,alpha =0.2)

x_hist.hist(x,bins=10,orientation='vertical',color ='grey')
```
