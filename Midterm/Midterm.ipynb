```python
Regression(X, y, grad_method, max_iter, alpha, tol, 
           decay, decay_iter, decay_rate, stop_delay_counter, 
           verbose, lam, poly, poly_deg)

class LinearRegression:

    def fit(self, X_train, y_train, X_test, y_test, algorithm="batch", 
            batch_size=64, alpha=0.001, max_iters=50000, tol=0.01):
        '''
        Use to fit (or train) linear regression model.
        X and y data must be split into train and test before call this method.
        But you don't need to perform data scaling and design matrix before 
        call this function. Everything will be done inside.
        Parameters:
            X_train : np.array(m, n)
            y_train : np.array(m, )
            X_test : np.array(m, n)
            y_test : np.array(m, )
            algorithm : str
                Refer to optimization algorithm (choose by yourself): 
                "batch" -> Batch Gradient Descent
                "stochastic" -> Stochastic Gradient Descent
                "mini-batch" -> Mini-Batch Gradient Descent                        
        '''
        # Attributes initialization
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test
        self.alpha = alpha
        self.max_iters = max_iters
        self.tol = tol
        self.loss_history = np.zeros(self.max_iters)
        self.algorithm = algorithm + " gradient descent"
        
        # Scaling data set (both X and y) with standardization
        self.feature_scaling()
        
        # Re-construct X_train into design matrix by insert column of 1 into
        # the first column of X_train
        self.X_train = np.insert(self.X_train, 0, 1, axis=1)
        
        # Find the shape of data then store them as object's attribute,
        # because we will use them again in the future.
        self.m, self.n = self.X_train.shape[:2]
        
        # We initialize theta (weight) here.
        # It will be used in optimization process.
        self.theta = np.ones(self.n)

        # Parameter optimization
        print("\nTraining ... ")
        if algorithm == "batch":
            self.batch_gradient_descent()
        elif algorithm == "stochastic":
            self.stochastic_gradient_descent()
        elif algorithm == "mini-batch":
            self.mini_batch_gradient_descent()
        else:
            print("In-valid optimization algorithm")
        return self.theta, self.loss_history

    def h(self, X, theta):
        return X.dot(theta)
    
    def cost_fn(self, y_pred):
        return (y_pred - self.y_train).T.dot(y_pred - self.y_train) / self.m
    
    def gradient(self, X_true, y_true, y_pred):
        return X_true.T.dot(y_pred - y_true)
    
    def early_break(self, i):
        '''
        Early break condition.
        When there is very low difference (thresholding by tol) in loss 
        between current and previous step. 
        '''
        if i > 0 and (abs(self.loss_history[i] - self.loss_history[i-1]) < self.tol):
            print("Early break in iteration : ", i)
            return True
        
    def batch_gradient_descent(self):
        '''
        Batch Gradient Descent parameter optimization algorithm.
        '''
        for i in range(self.max_iters):
            y_pred = self.h(self.X_train, self.theta)
            self.theta -= self.alpha * self.gradient(self.X_train, self.y_train, y_pred)
            self.loss_history[i] = self.cost_fn(y_pred)            
            if self.early_break(i):
                break
    
    def stochastic_gradient_descent(self):
        '''
        Stochastic Gradient Descent parameter optimization algorithm.
        '''
        for i in range(self.max_iters):
            idx = np.random.randint(self.m)
            X_i = self.X_train[idx, :]
            y_i = self.y_train[idx]
            y_pred = self.h(X_i, self.theta)
            # Prediction of all training samples
            y_pred_all = self.h(self.X_train, self.theta) 
            self.theta -= self.alpha * self.gradient(X_i, y_i, y_pred)
            self.loss_history[i] = self.cost_fn(y_pred_all)
            if self.early_break(i):
                break
            
    def mini_batch_gradient_descent(self):
        '''
        Mini-Batch Gradient Descent parameter optimization algorithm.
        '''
        for i in range(self.max_iters):
            idx = np.random.randint(self.m)        
            batch_X = self.X_train[:idx, :]
            batch_y = self.y_train[:idx]
            y_pred = self.h(batch_X, self.theta)
            y_pred_all = self.h(self.X_train, self.theta)
            self.theta -= self.alpha * self.gradient(batch_X, batch_y, y_pred)
            self.loss_history[i] = self.cost_fn(y_pred_all)
            if self.early_break(i):
                break
    
    def predict(self, X, theta):
        '''
        To predict y with input X and theta.
        It's not the same function as h(), because it's included with 
        data scaling and design matrix.
        Meaning, you don't need to perform data scaling and design matrix
        before call this method. 
        Just put your X and theta in to this method.
        * You cannot put your X if you nerver have girl/boy friend 
          or you have only one (current one).
        Parameters:
            X : np.array(m, n)
            theta : np.array(n+1, ) # n+1 because X isn't design matrix yet
        Return:
            Original scale of y_predict : np.array(m, )
        '''
        X = self.standardize(X, self.mean_X, self.std_X)
        X = np.insert(X, 0, 1, axis=1)
        y_pred = self.h(X, self.theta)
        return self.de_standardize(y_pred, self.mean_y, self.std_y)
    
    def plot_loss(self, fig=1):
        '''
        To plot loss of training (fitting) process.
        Just call it, don't need to worry about parameter and return.
        There is nothing.
        '''
        x = np.arange(self.max_iters)
        plt.figure(fig)        
        plt.plot(x, self.loss_history)
        plt.xlabel("iteration")
        plt.ylabel("loss")
        plt.title(self.algorithm)
```
